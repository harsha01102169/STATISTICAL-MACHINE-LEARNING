```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("/content/drive/MyDrive/Japan_life_expectancy.csv")
print(df)
```

```python
df.head()
```

```python
y = df['Life_expectancy']
Df = df.drop(columns = 'Prefecture',axis = 1)
Df.head()
```

```python
train, valid, test = np.split(Df.sample(frac=1), [int(0.6*len(df)), int(0.8*len(Df))])
```

```python
def scale_dataset(dataframe):
  X = dataframe[dataframe.columns[1:]].values
  y = dataframe[dataframe.columns[0]].values

  scaler = StandardScaler()
  X = scaler.fit_transform(X)

  data = np.hstack((X, np.reshape(y, (-1, 1))))

  return data, X, y
```

```python
train, X_train, y_train = scale_dataset(train)
valid, X_valid, y_valid = scale_dataset(valid)
test, X_test, y_test = scale_dataset(test)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
model = LinearRegression()

model.fit(X_train, y_train)

y_test_pred = model.predict(X_test)
y_train_pred = model.predict(X_train)
print(y_test_pred)
reg_train_error = mean_squared_error(y_train, y_train_pred)
reg_test_error = mean_squared_error(y_test, y_test_pred)
print("mean squared training error using linear regression:",reg_train_error)
print("mean squared testing error using linear regression:",reg_test_error)
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=0.1)
ridge.fit(X_train,y_train)
rid_train_pred = ridge.predict(X_train)
rid_test_pred = ridge.predict(X_test)
```

```python
rid_train_error = mean_squared_error(y_train, rid_train_pred)
rid_test_error = mean_squared_error(y_test, rid_test_pred)
print("mean squared training error using ridge regression:",rid_train_error)
print("mean squared testing error using ridge regression:",rid_test_error)
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.1)
lasso.fit(X_train,y_train)
las_train_pred = lasso.predict(X_train)
las_test_pred = lasso.predict(X_test)
```

```python
las_train_error = mean_squared_error(y_train, las_train_pred)
las_test_error = mean_squared_error(y_test, las_test_pred)
print("mean squared training error using lasso regression:",las_train_error)
print("mean squared testing error using lasso regression:",las_test_error)
```
